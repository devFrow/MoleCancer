```python
import os 
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
#carga de imagenes 

def carga_imagenes (ruta,etiqueta):
    imagenes = []
    etiquetas = []
    for x in os.listdir(ruta):
        if x.endswith('.jpg'):
            imagen = Image.open(os.path.join(ruta,x))
            imagenes.append(imagen)
            etiquetas.append(etiqueta)
    return imagenes,etiquetas

ruta_ent_benig = "C:\\Users\\nukke\\OneDrive\\Escritorio\\ds skin moles\\data\\train\\benign"  # directorio con las fotos de lunares benignos
ruta_ent_malig = "C:\\Users\\nukke\\OneDrive\\Escritorio\\ds skin moles\\data\\train\\malignant" #directorio con las fotos de lunares malignos

i_benig, et_benig = carga_imagenes(ruta_ent_benig,0)
i_malig, et_malig = carga_imagenes(ruta_ent_malig,1)
```

```python
# verifico si se cargaron las fotos 

# Visualizar una imagen benigna
plt.imshow(i_benig[0])
plt.title("Benigno")
plt.show()

# Visualizar una imagen maligna
plt.imshow(i_malig[0])
plt.title("Maligno")
plt.show()
```

```python
#redimencion de las imagenes 

tamaño = (224,224)
i_benig = [imagen.resize(tamaño) for imagen in i_benig]
i_malig = [imagen.resize(tamaño) for imagen in i_malig]

# Normalización de píxeles (escalar de 0-255 a 0-1)
i_benig = [np.array(imagen) / 255.0 for imagen in i_benig]
i_malig = [np.array(imagen) / 255.0 for imagen in i_malig]
```
